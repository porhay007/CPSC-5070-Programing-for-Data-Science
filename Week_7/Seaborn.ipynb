```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

plt.figure(figsize=(10,4))

# Without jitter
plt.subplot(1, 2, 1)
sns.stripplot(x="day", y="total_bill", data=tips, color="blue")
plt.title("Without jitter")

# With jitter
plt.subplot(1, 2, 2)
sns.stripplot(x="day", y="total_bill", data=tips, color="green", jitter=True)
plt.title("With jitter")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (6,8))
plt.show()
```

```python
sns.rugplot(tips['total_bill'] )
sns.kdeplot(tips['total_bill'] )
```

## Heatmap


```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
pvflights = flights.pivot_table(values ='passengers' ,index = 'month', columns = 'year' )
```

```python
pvflights
```

```python
sns.heatmap(pvflights, cmap = 'mako', linecolor = 'white', linewidth = 1)
```

```python
g = sns.PairGrid(tips, hue = 'smoker')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)

```

```python
g = sns.JointGrid(x = 'total_bill', y = 'tip', data = tips)
g.plot_marginals(sns.histplot, kde = True)
g.plot(sns.regplot, sns.histplot)
```
