```python
import pandas as pd
import numpy as np

# Number of rows
n = 500

df = pd.DataFrame({
    "Avg. Area Income": np.random.randint(40000, 80000, n),
    "Avg. Area House Age": np.random.randint(2, 15, n),
    "Avg. Area Number of Rooms": np.random.randint(3, 10, n),
    "Avg. Area Number of Bedrooms": np.random.randint(1, 6, n),
    "Area Population": np.random.randint(10000, 50000, n),
    "Price": np.random.randint(150000, 600000, n),
    "Address": [
        f"{np.random.randint(100,9999)} Random St, City {i}, State"
        for i in range(n)
    ]
})

df
```

### OCR : optical character recognition

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
digits
```

```python
digits.images[1]
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
```

```python
iso.fit(digits.data)
```

```python
data_projected = iso.transform(digits.data)
```

```python
data_projected.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_projected[:,0], data_projected[:,1], c = digits.target, cmap= plt.cm.get_cmap('nipy_spectral', 10), alpha=0.7, edgecolors='none')


plt.colorbar(label = 'digit label', ticks = range(10))
plt.clim(-0.5, 9.5)
```
