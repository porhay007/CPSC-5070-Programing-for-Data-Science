Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,10)
```

```python
x
```

```python
y = x **2
z = y **2
```

```python
y
```

```python
plt.plot(x,y,z, marker ='x', color = 'red')
plt.xlabel('X')
plt.ylabel('X ** 2')
plt.title('Sample matplotlib')
```

Sub plot

```python
plt.subplot(3,3,1)
plt.plot(x,y,'r--')
plt.subplot(3,3,9)
plt.plot(x,y,'k*-')
plt.subplot(3,3,5)
plt.plot(x,y,'go--')
```

OOP syntax (Object oriented programing) 

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.4,0.8]) # assuming the canvas (figure) is 1x1 [left, bottom, width, height]
axes.plot(x,y)
axes2 = fig.add_axes([0.6, 0.1, 0.4, 0.8])
axes2.plot(y,x)
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# 1. Initialize the figure
fig = plt.figure(figsize=(8, 6))

# 2. Define a 2x3 grid (2 rows, 3 columns) to divide the space
# We use a 2x3 grid to get the horizontal divisions for the left plots and the vertical division for the right plot.
gs = gridspec.GridSpec(2, 3, figure=fig)

# 3. Place the first subplot (Top-Left)
# It occupies the first row (0) and the first two columns (0:2).
ax1 = fig.add_subplot(gs[0, 0:2])

# 4. Place the second subplot (Bottom-Left)
# It occupies the second row (1) and the first two columns (0:2).
ax2 = fig.add_subplot(gs[1, 0:2])

# 5. Place the third subplot (Right Column)
# It occupies both rows (0:2) and the last column (2).
ax3 = fig.add_subplot(gs[0:2, 2])

# Adjust layout to prevent overlap
fig.tight_layout()

# Display the figure (optional, as the input image is already the result)
# plt.show()
```

```python
fig = plt.figure()
#axes = fig.add_axes([left, bottom, width, height])
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4,0.3])
axes3 = fig.add_axes([0.65, 0.2, 0.2,0.2])

axes1.plot(x,y, 'k')
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('my title')
axes2.plot(y,x, 'r')
axes2.set_xlabel('x mini label')
axes2.set_ylabel('y mini label')
axes2.set_ylabel('mini')
axes3.plot(x,x**3, 'go-')
```

```python
fig, axes = plt.subplots(1,2)

for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel('x label')
    ax.set_ylabel('y label')
    ax.set_title('my title')
    
fig.tight_layout() # use to clean space data
```

multiple curve in one plot and put legend to show label

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'x**2')
ax.plot(x,x**3, label= 'x**3')
ax.plot(x,x**2.5, label='x**2.5')

# Zoom the data
ax.set_xlim([1,4])
ax.set_ylim([0,20])

ax.legend(loc=(0.1,0.3)) # show the label in graph
```
