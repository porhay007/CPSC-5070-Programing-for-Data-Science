Common type of plot
+ Line : Trends
+ Scatter : Relationships
+ Bar : Comparisons
+ Histogram : Distributions

```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# This is the line plot to show trends

x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]
plt.plot(x,y,marker = "*", color = "red")
plt.xlabel("X")
plt.ylabel("X^2")
plt.grid(True)
plt.title("Squares")
plt.show()

```

```python
fig, ax = plt.subplots()
ax.plot(x,y, marker = "*", color = "crimson")
ax.set_title("Square (OO)")
ax.set_xlabel('X')
ax.set_ylabel('X ^ 2')
ax.grid(True)
plt.show()
```

Two subplot side

```python
fig, axes = plt.subplots(1,2, figsize= (8,3))
axes[0].plot([0,1,2,3],[0,1,4,9])
axes[1].plot([0,1,2,3],[0,1,8,27])
for ax in axes:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

plt.tight_layout
plt.show()
```

Sample line plot

```python
plt.plot([1,2,3,4,5],[3,2,1,5,8])
plt.title('Simple Plot')
plt.show()
```

Scatter plot

```python
import numpy as np

x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y, color= 'green')
plt.title('Scatter Plot Example')
plt.show()
```

Bar Chart Example

```python
categories = ['A', 'B', 'C']
values = [5,2,7]
plt.bar(categories, values, color = 'teal')
plt.title('Bar Chart Example')
plt.show()
```

Historgram Example

```python
data = np.random.randn(200)
plt.hist(data, bins=20, edgecolor= 'red', color= 'Orange')
plt.title('Histogram example')
plt.show()
```

Dashboard Example

```python
fig, axes = plt.subplots(2,2,figsize= (8,6))
axes = axes.ravel()
axes[0].plot(x,y); axes[0].set_title('Line')
axes[1].scatter(x,y); axes[1].set_title('scatter')
axes[2].bar(categories, values); axes[2].set_title('Bar')
axes[3].hist(data,bins =20); axes[3].set_title('Hist')

for ax in axes: ax.grid(True)
fig.suptitle("Dashboard Example", y = 1.03)
plt.show()
```
